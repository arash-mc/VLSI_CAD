## 2- Define a function With truth table

```python
X = exprvars('x', 3)
f = truthtable(X, "00000001")
print(f)
```

```python
truthtable2expr(f)
```

## BDDs

```python
bdd = expr2bdd(f)
Source(bdd.to_dot())
```
